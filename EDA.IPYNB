```python
tips.plot(kind='scatter', x='day',y='size');
plt.show
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(tips, hue="sex", height=4) \
   .map(plt.scatter, "day", "size") \
   .add_legend();
plt.show();
```

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(tips, hue="sex", height=3);
plt.show()
```

```python
import numpy as np
tips_sex = tips.loc[tips["sex"] == "sex"];
tips_day = tips.loc[tips["sex"] == "day"];
tips_time = tips.loc[tips["sex"] == "time"];
plt.plot(tips_sex["sex"], np.zeros_like(tips_sex['sex']), '2')
plt.plot(tips_day["day"], np.zeros_like(tips_day['day']), '2')
plt.plot(tips_time["time"], np.zeros_like(tips_time['time']), '2')

plt.show()
```

```python
sns.FacetGrid(tips, hue="sex", height=5) \
   .map(sns.distplot, "size") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(tips, hue="total_bill", height=5) \
   .map(sns.distplot, "sex") \
   .add_legend();
plt.show();
```
